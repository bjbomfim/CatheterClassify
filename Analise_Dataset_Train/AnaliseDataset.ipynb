```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
raio_x_invasive_proce  = pd.read_csv('../Dataset/train.csv')
```

```python
raio_x_invasive_proce.info()
```

```python

ett = ['ETT - Abnormal', 'ETT - Borderline', 'ETT - Normal']
ngt = ['NGT - Abnormal', 'NGT - Borderline', 'NGT - Incompletely Imaged', 'NGT - Normal']
cvc = ['CVC - Abnormal', 'CVC - Borderline', 'CVC - Normal']
```

```python
plt.hist(x = raio_x_invasive_proce[ett].where(raio_x_invasive_proce[ett] == 1), label=ett);

plt.legend(loc='upper right')

plt.title('Histograma de Tubo Endotraqueal')
plt.ylabel('Frequência')
plt.show();
```

```python
plt.hist(x = raio_x_invasive_proce[ngt].where(raio_x_invasive_proce[ngt] == 1), label=ngt);

plt.legend(loc='upper right')

plt.title('Histograma de Sonda Nasogástrica')
plt.ylabel('Frequência')
plt.show();
```

```python
plt.hist(x = raio_x_invasive_proce[cvc].where(raio_x_invasive_proce[cvc] == 1), label=cvc);

plt.legend(loc='upper right')

plt.title('Histograma de Cateter Venoso Central')
plt.ylabel('Frequência')
plt.show();
```

```python
plt.hist(x = raio_x_invasive_proce['Swan Ganz Catheter Present'].where(raio_x_invasive_proce['Swan Ganz Catheter Present'] == 1), label='Swan Ganz Catheter Present');

plt.legend(loc='upper right')

plt.title('Histograma de Cateter Venoso Central')
plt.ylabel('Frequência')
plt.show();
```

```python
raio_x_invasive_proce.where(raio_x_invasive_proce == 1).count()
```

```python
porcent = np.divide(raio_x_invasive_proce.where(raio_x_invasive_proce == 1).count(), np.sum(raio_x_invasive_proce.where(raio_x_invasive_proce == 1 ).count()))

listas = [ett, ngt, cvc]
x = []
for lista in listas:
    
    x.extend(lista)
```

```python

    
x.append("Swan Ganz Catheter Present")
plt.figure(figsize=(10, 6))
plt.barh(x, porcent[x]*100, color=['skyblue', 'skyblue', 'skyblue',  'lightgreen',  'lightgreen',  'lightgreen',  'lightgreen',  'gold', 'gold', 'gold', 'lightcoral',])

plt.xlabel('Porcentagem')
plt.title('Gráfico dos Grupos')

plt.show()
```

```python
porcent = np.divide(raio_x_invasive_proce.where(raio_x_invasive_proce[ett] == 1).count(), np.sum(raio_x_invasive_proce.where(raio_x_invasive_proce[ett] == 1 ).count()))
porcent_ett = porcent[ett]

plt.figure(figsize=(10, 6))
plt.barh(ett, porcent_ett*100)

plt.xlabel('Porcentagem')
plt.title('Gráfico do Grupo ETT')

plt.show()
```

```python
porcent = np.divide(raio_x_invasive_proce.where(raio_x_invasive_proce[ngt] == 1).count(), np.sum(raio_x_invasive_proce.where(raio_x_invasive_proce[ngt] == 1 ).count()))
porcent_ngt = porcent[ngt]

plt.figure(figsize=(10, 6))
plt.barh(ngt, porcent_ngt*100)

plt.xlabel('Porcentagem')
plt.title('Gráfico do Grupo NGT')

plt.show()
```

```python
porcent = np.divide(raio_x_invasive_proce.where(raio_x_invasive_proce[cvc] == 1).count(), np.sum(raio_x_invasive_proce.where(raio_x_invasive_proce[cvc] == 1 ).count()))
porcent_cvc = porcent[cvc]

plt.figure(figsize=(10, 6))
plt.barh(cvc, porcent_cvc*100)

plt.xlabel('Porcentagem')
plt.title('Gráfico do Grupo cvc')

plt.show()
```
